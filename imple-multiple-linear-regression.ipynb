Scale prediction

```python
#scale prediction 
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df = pandas.read_csv('/kaggle/input/multiple-linear-regression-dataset/multiple_linear_regression_dataset.csv')

X = df[['age', 'experience']]
y = df['income']

scaledX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])

predicted = regr.predict([scaled[0]])
print(f'predicted scale is :{predicted}')
```

Visualisation

```python
import matplotlib.pyplot as plt

# Predict the values
y_pred = regr.predict(X)

# Scatter plot of actual vs. predicted values
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2)
plt.xlabel('Actual Income')
plt.ylabel('Predicted Income')
plt.title('Actual vs Predicted Income')
plt.show()
```

3D visualisation

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['age'], df['experience'], y, color='blue')
ax.scatter(df['age'], df['experience'], y_pred, color='red')

ax.set_xlabel('Age')
ax.set_ylabel('Experience')
ax.set_zlabel('Income')
plt.title('3D Plot of Age, Experience, and Income')
plt.show()
```
