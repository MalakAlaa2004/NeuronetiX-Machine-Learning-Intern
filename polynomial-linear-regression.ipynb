```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
df = pd.read_csv('/kaggle/input/polynomial-linear-regression-dataset/rank_salary.csv')
x = df['Level']
y = df['Salary']
mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(min(x),max(x), 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Salary Rank')
plt.show()
r2_value=r2_score(y,mymodel(x))
predicted=mymodel(10)
print(f'r2 value is :{r2_value}')
print(f'predicted value :{predicted}')
```
