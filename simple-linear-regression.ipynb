```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas
df = pandas.read_csv('/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv')
x=df['YearsExperience']
y=df['Salary']
slope , intercept , r , p , std_err = stats.linregress(x,y)

def myfun(x):
    return slope * x + intercept 
mymodel = list(map(myfun,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.xlabel('Years of experience ')
plt.ylabel('Salary')
plt.title('Years of experience VS Salary')
plt.show()
predicted = myfun(10)
print(f"Correlation coefficient (r): {r}")
print(f"the predicted salary after 10 years of experience is:{predicted}" )
```
